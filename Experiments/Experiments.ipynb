```python
import random
df.size
flat_list = df.values.flatten()


my_set = set(flat_list)
unique_list = list(my_set)


```

```python
def jaccard_similarity(A, B):
    #Check if a and b are sets if not convert them to sets
    if not isinstance(A, set):
        A = set(A)
    if not isinstance(B, set):
        B = set(B)
    #Find intersection of two sets
    nominator = A.intersection(B)

    #Find union of two sets
    denominator = A.union(B)

    #Take the ratio of sizes
    similarity = len(nominator)/len(denominator)

    return similarity

```

```python
# importing the required module
import matplotlib.pyplot as plt

# x axis values
x = [0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.5,0.6,0.7,0.8,0.9,1]
x = [int(len(unique_list) * i)  for i in x]
l =  []
y = []
for i in x:
    l.append(random.sample(unique_list, i))

# x = [random.sample(unique_list, ]
# y = []

for i in l:
    y.append(jaccard_similarity(unique_list, i))


#plotting the points
plt.plot(x, y)

# # naming the x axis
plt.xlabel('x - axis')
# # naming the y axis
plt.ylabel('y - axis')

# # giving a title to my graph
plt.title('!')

# # function to show the plot
plt.show()
```
